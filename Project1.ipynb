```python
#Task8 Comment on distribution of the match's duration. Use seaborn.
a=sns.distplot( data['matchDuration'] )
print(a)
plt.title("Task-8")
plt.xticks(rotation=70);
plt.show()
print()
```

```python
#Task9 Comment on distribution of the walk distance. Use seaborn. 
print(sns.distplot( data['walkDistance'] ))
plt.title("task-9")
plt.xticks(rotation=70);
plt.show()
print()
```

```python
#Task10 Plot distribution of the match's duration vs walk distance one below the other.
plt.style.use('classic')
plt.figure()
plt.subplot(2,1,1)
plt.plot(data["matchDuration"],"+")
print(plt.xlabel("Match Duration"))
plt.subplot(2,1,2)
plt.plot(data["walkDistance"],"--")
print(plt.xlabel("Walk Distance"))
plt.title("task-10")
plt.xticks(rotation=70);
plt.show()
print()
```

```python
#Task11 Plot distribution of the match's duration vs walk distance side by side.
plt.style.use('classic')
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(data["matchDuration"])
print(plt.xlabel("Match Duration"))
plt.subplot(1,2,2)
plt.plot(data["walkDistance"])
print(plt.xlabel("Walk Distance"))
plt.title("Task-11")
plt.xticks(rotation=70);
plt.show()
```
